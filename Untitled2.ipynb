```python
import torch
from torchvision import models, transforms
from PIL import Image
import os
import numpy as np

resnet = models.resnet50(pretrained=True)
resnet = torch.nn.Sequential(*list(resnet.children())[:-1])  
resnet.eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

def extract_features(image_path, model, transform):
    img = Image.open(image_path).convert('RGB')
    img = transform(img).unsqueeze(0)  
    with torch.no_grad():
        features = model(img)
    return features.squeeze().cpu().numpy() 
unlabelled_images_dir = 'image'

valid_image_extensions = ['.jpg', '.jpeg', '.png']
image_paths = [os.path.join(unlabelled_images_dir, img) for img in os.listdir(unlabelled_images_dir)
               if img.endswith(tuple(valid_image_extensions)) and not img.startswith('.')]

image_features = []

for img_path in image_paths:
    try:
        features = extract_features(img_path, resnet, transform)
        image_features.append(features)
    except Exception as e:
        print(f"Error processing image {img_path}: {e}")

image_features = np.array(image_features)
```

```python
import os
import cv2
import numpy as np
import torch
from torchvision import models, transforms
from PIL import Image
from sklearn.cluster import KMeans

# Load pre-trained ResNet (remove final fully connected layer for feature extraction)
resnet = models.resnet50(pretrained=True)
resnet = torch.nn.Sequential(*list(resnet.children())[:-1])  # Remove the last FC layer
resnet.eval()

# Define image transformation (to match ResNet input requirements)
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Feature extraction function (ResNet)
def extract_features(image_path, model, transform):
    if not os.path.exists(image_path):
        print(f"File not found: {image_path}")
        return None
    img = Image.open(image_path).convert('RGB')
    img = transform(img).unsqueeze(0)  # Add batch dimension
    with torch.no_grad():
        features = model(img)
    return features.squeeze().cpu().numpy()  # Return as numpy array

# Color histogram extraction function
def extract_color_histogram(image_path):
    if not os.path.exists(image_path):
        print(f"File not found: {image_path}")
        return None
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))
    hist = cv2.calcHist([img], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    return cv2.normalize(hist, hist).flatten()

# Directory of unlabelled images (with subfolders severe, mild, moderate)
unlabelled_images_dir = 'image'

# Gather all image paths from subdirectories (severe, mild, moderate)
valid_image_extensions = ['.jpg', '.jpeg', '.png']
image_paths = []

for root, dirs, files in os.walk(unlabelled_images_dir):
    for file in files:
        if file.endswith(tuple(valid_image_extensions)) and not file.startswith('.'):
            image_paths.append(os.path.join(root, file))

# Extract combined features (ResNet + Color Histogram)
combined_features = []

for img_path in image_paths:
    # Extract ResNet features
    resnet_features = extract_features(img_path, resnet, transform)
    
    if resnet_features is None:
        continue  # Skip this image if it wasn't found
    
    # Extract color histogram features
    color_hist_features = extract_color_histogram(img_path)
    
    if color_hist_features is None:
        continue  # Skip this image if it wasn't found
    
    # Combine both feature sets
    combined_feature = np.concatenate((resnet_features, color_hist_features))
    combined_features.append(combined_feature)

# Convert to numpy array
combined_features = np.array(combined_features)

# Cluster using combined features
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(combined_features)
cluster_labels = kmeans.labels_

print("Clustering complete!")
```

```python
import numpy as np
import os
import cv2
import torch
from torchvision import models, transforms
from PIL import Image
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, average_precision_score
from sklearn.preprocessing import LabelEncoder

# Load pre-trained ResNet (remove final fully connected layer for feature extraction)
resnet = models.resnet50(pretrained=True)
resnet = torch.nn.Sequential(*list(resnet.children())[:-1])  # Remove the last FC layer
resnet.eval()

# Define image transformation (to match ResNet input requirements)
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Feature extraction function (ResNet)
def extract_features(image_path, model, transform):
    img = Image.open(image_path).convert('RGB')
    img = transform(img).unsqueeze(0)  # Add batch dimension
    with torch.no_grad():
        features = model(img)
    return features.squeeze().cpu().numpy()  # Return as numpy array

# Color histogram extraction function
def extract_color_histogram(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))
    hist = cv2.calcHist([img], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    return cv2.normalize(hist, hist).flatten()

# Directory of images (with subfolders severe, mild, moderate)
image_dir = 'image'

# Gather all image paths and labels
image_paths = []
labels = []

for root, dirs, files in os.walk(image_dir):
    for file in files:
        if file.endswith(('.jpg', '.jpeg', '.png')):
            image_paths.append(os.path.join(root, file))
            # Label based on folder name
            if 'severe' in root:
                labels.append('severe')
            elif 'mild' in root:
                labels.append('mild')
            elif 'moderate' in root:
                labels.append('moderate')

# Encode labels as numeric values
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)

# Extract combined features (ResNet + Color Histogram)
combined_features = []

for img_path in image_paths:
    # Extract ResNet features
    resnet_features = extract_features(img_path, resnet, transform)
    
    # Extract color histogram features
    color_hist_features = extract_color_histogram(img_path)
    
    # Combine both feature sets
    combined_feature = np.concatenate((resnet_features, color_hist_features))
    combined_features.append(combined_feature)

combined_features = np.array(combined_features)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(combined_features, labels_encoded, test_size=0.3, random_state=42)

# Train an SVM classifier
svm_classifier = SVC(kernel='linear', probability=True, random_state=42)
svm_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = svm_classifier.predict(X_test)
y_pred_proba = svm_classifier.predict_proba(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')

# Calculate mAP (mean Average Precision)
mAP = average_precision_score(np.eye(3)[y_test], y_pred_proba, average='macro')

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Mean Average Precision (mAP): {mAP:.4f}")
```
